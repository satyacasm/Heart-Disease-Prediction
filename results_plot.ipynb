```python
# results_plots.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the results data
results_df = pd.read_csv('model_results.csv')

# Set the style for the plots
sns.set(style="whitegrid")

# Plot the accuracy of the models
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Accuracy', data=results_df, palette='viridis')
plt.title('Model Accuracy Comparison')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot the precision of the models
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Precision', data=results_df, palette='plasma')
plt.title('Model Precision Comparison')
plt.xlabel('Model')
plt.ylabel('Precision')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot the recall of the models
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Recall', data=results_df, palette='inferno')
plt.title('Model Recall Comparison')
plt.xlabel('Model')
plt.ylabel('Recall')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot the F1-Score of the models
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='F1-Score', data=results_df, palette='magma')
plt.title('Model F1-Score Comparison')
plt.xlabel('Model')
plt.ylabel('F1-Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create a heatmap of the results
plt.figure(figsize=(10, 6))
pivot_df = results_df.melt(id_vars='Model', var_name='Metric', value_name='Value')
pivot_table = pivot_df.pivot(index='Model', columns='Metric', values='Value')
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title('Model Performance Heatmap')
plt.xlabel('Metric')
plt.ylabel('Model')
plt.tight_layout()
plt.show()
```
